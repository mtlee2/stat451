```python
# Data
labels = ["Training Accuracy", "Testing Accuracy"]
models = ['Decision Tree', 'Random Forest', "Logistic Regression"]
v0 = [0.99, 0.56]  # Decision Tree
v1 = [1.0, 0.58]   # Random Forest
v2 = [0.57, 0.53]  # Logistic Regression

# Parameters for bar width and positions
bar_width = 0.25
x_pos = np.arange(len(labels))  # Create positions for training and testing

# Create the figure and axes
fig, ax = plt.subplots(figsize=(8, 6))

# Plot each model's bars with appropriate offsets
rects1 = ax.bar(x_pos - bar_width, v0, bar_width, label=models[0])
rects2 = ax.bar(x_pos, v1, bar_width, label=models[1])
rects3 = ax.bar(x_pos + bar_width, v2, bar_width, label=models[2])

# Customizing the plot
ax.set_xticks(x_pos)  # Set x-axis ticks to match the group positions
ax.set_xticklabels(labels)  # Label x-axis ticks with "Training" and "Testing"
ax.set_ylabel("Accuracy")
ax.set_ylim(0.0, 1.2)  # Extend y-axis for readability
ax.set_title("Model Comparison: Training vs Testing Accuracy")
ax.legend()

# Adding values on top of the bars
for rects in [rects1, rects2, rects3]:
    for rect in rects:
        height = rect.get_height()
        ax.text(
            rect.get_x() + rect.get_width() / 2,
            height + 0.02,
            f"{height:.2f}",
            ha='center',
            va='bottom'
        )

# Show the plot
plt.show()
```
